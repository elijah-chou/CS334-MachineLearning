```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("eng_xTest.csv")
target = pd.read_csv("eng_yTest.csv")
```

```python
datetime = pd.to_datetime(df["date"])
df["datetime"] = datetime
df['day'] = df["datetime"].dt.day
df = df.drop(columns=['date', 'datetime'])
df["target"] = target
```

```python
correlations = df.corr()
```

```python
sns_heatmap = sns.heatmap(correlations)
plt.show()
```

```python
fig = sns_heatmap.get_figure()
fig.savefig('correlation_heatmap.png')
```
